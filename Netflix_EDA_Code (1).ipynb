```python
#Tv Shows got rating “R” after year 2018? 
nfdata.dropna(inplace=True)
nfdata["Year"]=nfdata["Year"].astype(int)
Ratings=nfdata[(nfdata["Rating"]=="R" ) & (nfdata["Year"]>=2018)][["Category","Year"]]
print("No of shows that got 'R' rating after 2018 :",len(Ratings))
```
